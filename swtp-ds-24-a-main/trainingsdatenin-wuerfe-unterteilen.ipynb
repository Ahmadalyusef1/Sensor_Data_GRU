```python
import pandas as pd
```

```python

# Daten einlesen
combined_df = pd.read_csv('data/combined_data_mit_ruhezustaende')

# Neue Spalte f체r Wurfnummer hinzuf체gen
combined_df['wurf_nummer'] = 0

# Initiale Wurfnummer
current_wurf_nummer = 1
in_resting_state = False

# Iteriere durch die Datenpunkte
for i in range(len(combined_df)):
    if combined_df['label'].iloc[i] == 2:
        in_resting_state = True
    else:
        if in_resting_state:
            current_wurf_nummer += 1
            in_resting_state = False
        combined_df.at[i, 'wurf_nummer'] = current_wurf_nummer

# Speichern des aktualisierten DataFrames als CSV
output_path = 'data/combined_df_with_wurfnummer.csv'
combined_df.to_csv(output_path, index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Daten einlesen
combined_df = pd.read_csv('data/combined_df_with_wurfnummer.csv')  # Datei anpassen

# Auswahl der ersten 6000 Datenpunkte
subset_df = combined_df.head(6000)

# Diagramm erstellen
plt.figure(figsize=(14, 7))
plt.scatter(subset_df['timestamp'], subset_df['label'], c=subset_df['label'], cmap='viridis', marker='o', alpha=0.5)
plt.colorbar(label='Label')
plt.xlabel('Timestamp')
plt.ylabel('Label')
plt.title('Labels in Abh채ngigkeit von Timestamp f체r die ersten 6000 Datenpunkte')
plt.grid(True)

# Setzen der Y-Achsen-Ticks auf die Labels 0, 1, und 2
plt.yticks([0, 1, 2], ['Normal (0)', 'Anomalie (1)', 'Ruhezustand (2)'])

plt.show()
```
