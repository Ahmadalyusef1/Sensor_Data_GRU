```python
import matplotlib.pyplot as plt
import pandas as pd

# Daten aus der CSV-Datei laden
daten = pd.read_csv('data/normal_data.csv')

# Histogramme für jede Variable zeichnen
plt.figure(figsize=(18, 12))

# Histogramm für 'ax'
plt.subplot(3, 2, 1)
plt.hist(daten['ax'], bins=30, color='blue', edgecolor='black')
plt.title('Histogramm von ax')
plt.grid(True)
# Verbesserung der x-Achsen-Beschriftung
plt.xticks(rotation='vertical')

# Histogramm für 'ay'
plt.subplot(3, 2, 2)
plt.hist(daten['ay'], bins=30, color='green', edgecolor='black')
plt.title('Histogramm von ay')
plt.grid(True)
plt.xticks(rotation='vertical')

# Histogramm für 'az'
plt.subplot(3, 2, 3)
plt.hist(daten['az'], bins=30, color='red', edgecolor='black')
plt.title('Histogramm von az')
plt.grid(True)
plt.xticks(rotation='vertical')


# Layout anpassen und anzeigen
plt.tight_layout()
plt.show()
```

```python
import pandas as pd


anomalie_daten = pd.read_csv('data/anomaly_data.csv')
normal_daten = pd.read_csv('data/normal_data.csv')

obergrenze = 0.5
untergrenze = -0.5


stabile_anomalie_bereiche = anomalie_daten[
    (anomalie_daten['ax'].between(untergrenze, obergrenze)) &
    (anomalie_daten['ay'].between(untergrenze, obergrenze)) &
    (anomalie_daten['az'].between(untergrenze, obergrenze))
]


anomalie_daten.loc[stabile_anomalie_bereiche.index, 'label'] = 2

stabile_normal_bereiche = normal_daten[
    (normal_daten['ax'].between(untergrenze, obergrenze)) &
    (normal_daten['ay'].between(untergrenze, obergrenze)) &
    (normal_daten['az'].between(untergrenze, obergrenze))
]

normal_daten.loc[stabile_normal_bereiche.index, 'label'] = 2

kombinierte_daten = pd.concat([anomalie_daten, normal_daten])

kombinierte_daten.to_csv('data/combined_data_with_resting_state.csv', index=False)

print(f"Anzahl der stabilen Blöcke in Anomalie-Daten: {stabile_anomalie_bereiche.shape[0]}")
print(f"Anzahl der stabilen Blöcke in Normal-Daten: {stabile_normal_bereiche.shape[0]}")
print(f"Gesamtanzahl der stabilen Blöcke: {stabile_anomalie_bereiche.shape[0] + stabile_normal_bereiche.shape[0]}")
```
